```python
#_**OpenCV to convert image from a BGR channel to RGB then to Gray Scale**_
```

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
img =cv2.imread("lena.jpg")
```

```python
plt.imshow(img)
```

```python
## Convert the image to RGB from BGR
```

```python
img=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(img)
```

```python
#Convert image to a Gray Scale Imgae
```

```python
gray_img = cv2.cvtColor(img,cv2.COLOR_RGB2GRAY)
```

```python
plt.imshow(gray_img,cmap='gray')
```
